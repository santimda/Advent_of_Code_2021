```python
# Read the puzzle, it contains a list of syntax lines
#with open('syntax_example.txt') as f:
with open('syntax.txt') as f:
    lines = [x.strip('\n') for x in f.readlines()]
print(lines)
```

```python
# Valid closers:
closers = {')', ']', '>', '}'}
# Corresponding openers
def opener(char):
    if char == ')':
        op = '('
    elif char == ']':
        op = '['
    elif char == '}':
        op = '{'
    elif char == '>':
        op = '<'
    return(op)
```

```python
def count_chunk(chunk):
    np, nb, nc, na = 0, 0, 0, 0
    for char in chunk:
        if char == '(':
            np += 1
        elif char == ')':
            np -= 1
        elif char == '[':
            nb += 1
        elif char == ']':
            nb -= 1
        elif char == '{':
            nc += 1
        elif char == '}':
            nc -= 1
        elif char == '<':
            na += 1
        elif char == '>':
            na -= 1
    return (np, nb, nc, na)

def check_chunk(chunk):
    if sum(count_chunk(chunk)) != 0:
        valid = False
    else:
        valid = True
    return valid      

def obtain_chunk(string):
    cl = string[-1] # The last characters is the closer
    op = opener(cl)
    n = 1 # n = number of closers - number of openers (chunk closes when n=0)
    l = 0
    while l < len(string)-1 and n != 0:
        l += 1
        if string[-1-l] == cl:
            n += 1
        elif string[-1-l] == op:
            n -= 1
        #print(f'{string[-1-l:]}, {l=}, {n=}')
    return string[-1-l:]
        
    if sum(count_chunk(chunk)) != 0:
        valid = False
    else:
        valid = True
    return valid            
```

```python
# Check for each line if it is corrupted
# characters: p=(), b=[], c={}, a=<>
bad_chars = []
for i, line in enumerate(lines):
    # define counters of how many of the characters occurred in each line
    chunk = ''
    valid = True
    for j, char in enumerate(line):
        if valid: 
            # check if the char closes a chunk and obtain that chunk
            if char in closers:
                chunk = obtain_chunk(line[:j+1])
                #print(f'line {i+1},  {j = } ,  {chunk = }')
                # check if the closed chunk produced an error
                valid = check_chunk(chunk)
                if not valid: 
                    print(f'ERROR in line {i+1} for {chunk = }, bad character {char}')
                    bad_chars.append(char)
            #print(f'line {i+1} {line[:j+1]} , {chunk = } {count_chunk(chunk)}')
```
