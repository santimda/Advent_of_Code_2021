## First part

```python
# Read the puzzle, it contains a list of start and end points
lines_raw = []
#with open('vents_example.txt') as f:
with open('vents.txt') as f:
    lines_raw = f.readlines()

# Save start points and end points in p_is and p_fs
# Convert from strings to integers
pi_s = []
pf_s = []
for line in lines_raw:
    p1, p2 = line.strip(' ').split('->')
    pi = [int(i) for i in p1.split(',')]
    pf = [int(i) for i in p2.split(',')]
    pi_s.append(pi)
    pf_s.append(pf)
    
print(pi_s)
print(pf_s)
```

```python
# Make a grid of points, but first determine the furthermost corners

# Assign seed values using the first point
#x_min, x_max, y_min, y_max = pi_s[0][0], pi_s[0][0], pi_s[0][1], pi_s[0][1]
x_max, y_max = pi_s[0][0], pi_s[0][1]

zip_object = zip(pi_s, pf_s)
for pi, pf in zip_object:
#    x_min = min( x_min, pi[0], pf[0] )
    x_max = max( x_max, pi[0], pf[0] )
#    y_min = min( y_min, pi[1], pf[1] )
    y_max = max( y_max, pi[1], pf[1] )

#print(x_min, x_max)
#print(y_min, y_max)
print(x_max, y_max)
```

```python
#positions = np.zeros( (x_max-x_min+1, y_max-y_min+1) )
positions = np.zeros( (x_max+1, y_max+1), dtype = np.uint32 )
zip_object = zip(pi_s, pf_s)
for pi, pf in zip_object:
    #Search for vertical lines
    if pi[0] == pf[0]:
        #create all the points in the segment
        x = pi[0] #- x_min
        yi = min(pi[1], pf[1])
        yf = max(pi[1], pf[1])
        for y in range(yi, yf+1):
            positions[y, x] += 1
            #print(pi, pf, x, y)

    #Search for horizontal lines
    elif pi[1] == pf[1]: 
        #create all the points in the segment
        y = pi[1] #- y_min
        xi = min(pi[0], pf[0])
        xf = max(pi[0], pf[0])
        for x in range(xi, xf+1):
            positions[y, x] += 1
            #print(pi, pf, x, y)

print(positions)
```

```python
#Check in which positions the value is > 1 (i.e, not 0 and not 1)
non_zeros = np.count_nonzero(positions)
ones = np.count_nonzero(positions == 1)
print('Answer: ' + str(non_zeros - ones) )
```
